```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.ticker import StrMethodFormatter
```

```python
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.size": 9,
    "axes.titlesize": 11,
    "axes.labelsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 10,
    "figure.dpi": 300,
    "lines.linewidth": 1.5,
    "savefig.dpi": 600,
    "axes.spines.top": False,
    "axes.spines.right": False,
})
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['mathtext.rm'] = 'stix'
plt.rcParams['mathtext.it'] = 'stix'
plt.rcParams['mathtext.bf'] = 'stix'
```

```python
latent_features=['OH', 'CH3OH', 'CH3']
```

```python
permutation_scores=pd.read_csv('permutation_importance_scores_phi8.csv')
permutation_scores=pd.DataFrame(permutation_scores)
```

```python
carrier_1=permutation_scores.iloc[:,1].to_numpy()
carrier_2=permutation_scores.iloc[:,2].to_numpy()
carrier_3=permutation_scores.iloc[:,3].to_numpy()
```

```python
fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(projection='3d')
pl=ax.scatter(carrier_2, carrier_3, carrier_1, c=carrier_1, cmap='Spectral', s=150)
cbar=plt.colorbar(pl, shrink=0.5)
cbar.ax.tick_params(labelsize=15)
cbar.ax.set_title(r'$i_{OH}$',fontsize=30)
ax.set_xlabel(r'$i_{CH3OH}$',fontsize=30,labelpad=25)
ax.set_ylabel(r'$i_{CH3}$',fontsize=30, labelpad=25)
ax.zaxis.set_rotate_label(False)
ax.set_zlabel(r'$i_{OH}$',fontsize=30, rotation=90,labelpad=15)
ax.tick_params(axis="x", labelsize=20)
ax.tick_params(axis="y", labelsize=20)
ax.tick_params(axis="z", labelsize=20)
ax.view_init(azim=45)
ax.dist=15
ax.set_box_aspect(aspect=None, zoom=0.8)

a = plt.axes([0.57, .52, .20, .15])
plt.scatter(carrier_3, carrier_2, c=carrier_1,cmap='Spectral',s=50)
plt.ylabel(r'$i_{CH3OH}$', fontsize=25,rotation='horizontal',loc='top')
plt.xlabel(r'$i_{CH3}$', fontsize=25, loc='right')
#plt.title('Probability')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}'))
plt.savefig('grimech_sensitivities_3d_view.png', dpi=600)
plt.show()
```
